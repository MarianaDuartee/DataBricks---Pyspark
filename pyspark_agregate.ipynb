```
import pyspark.sql.functions as F
```

```
dados = [
         ("Anderson","Vendas","SP",1500.00,34,1000.00),
         ("Kennedy","Vendas","CE",1200.00,56,2000.00),
         ("Bruno","Vendas","SP",1100.00,30,2300.00),
         ("Maria","Financas","CE",3600.00,24,2300.00),
         ("Eduardo","Financas","CE",4500.00,40,2400.00),
         ("Mendes","Financas","RS",8000.00,36,1900.00),
         ("Kethlyn","Financas","RS",1200.00,53,1500.00),
         ("Thiago","Marketing","GO",1100.00,25,1800.00),
         ("Carla","Marketing","GO",2600.00,50,2100.00)
]

schema = ["nome","departamento","estado","salario","idade","bonus"]

df = spark.createDataFrame(data=dados,schema=schema)

df.printSchema()
#df.show() ou
display(df)
```

Exibe quantos valores distintos tem em um grupo dentro de um DataFrame

```
approx = df.select(F.approx_count_distinct('salario')).collect()[0][0]
print(f'approx_count_distinct: {approx}')
```

Retorna a media de valores atraves da coluna selecionada

```
avg = df.select(F.avg('salario')).collect()[0][0]
print(f'avg:{avg}')
```

Pega todos os valores de uma coluna e coloca dentro de uma lista

```
df.select(F.collect_list('salario')).display()
```

Retorna uma lista dos valores distintos de uma coluna

```
df.select(F.collect_set('salario')).show(truncate=False)
```

Retorna o numero de elementos distintos

Combinando duas ou mais colunas, analisa os elementos distintos em ambas as colunas ao mesmo tempo.

No exemplo abaixo: analisa os salarios distintos POR departamento

```
df.select(F.countDistinct(F.col('departamento'),F.col('salario'))).display()
df.show()
```

Retorna o primeiro | ultimo elemento da coluna

```
df.select(F.first(F.col("nome"))).display()
df.select(F.last(F.col('nome'))).display()
```

Retorna o maior|menor valor |soma dos valores | media entre os valores

```
df.select(F.max('salario')).display()
df.select(F.min('salario')).display()
df.select(F.sum('salario')).display()
df.select(F.mean('salario')).display()
```
